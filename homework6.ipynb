```python
import numpy as np
import matplotlib.pyplot as plt


f = lambda x: -2 * x

def RK4(t, y0):
    """
    四阶 Runge-Kutta
    """
    n = len(t)
    y = np.zeros(n)
    y[0] = y0
    for i in range(0, n-1):
        h = t[i+1] - t[i]   
        k1 = f(y[i])
        k2 = f(y[i] + h/2*k1) 
        k3 = f(y[i] + h/2*k2)
        k4 = f(y[i] + h*k3)
        y[i+1] = y[i] + h/6 * (k1 + 2*k2 + 2*k3 + k4)
    return y

t = np.linspace(0, 10, 200)
y = RK4(t, y0=2)

plt.plot(t, y, label = 'RK4')
plt.plot(t, np.exp(-2*t) * 2, label = 'true')
plt.legend()
plt.show()
```
